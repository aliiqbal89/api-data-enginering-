```python
import requests
import csv

response = requests.get("https://restcountries.com/v3.1/all")
countries = response.json()

data_rows = []

for country in countries:
    row = {
        "Name": country.get("name", {}).get("common"),
        "Official Name": country.get("name", {}).get("official"),
        "Capital": ', '.join(country.get("capital", [])) if country.get("capital") else None,
        "Region": country.get("region"),
        "Subregion": country.get("subregion"),
        "Population": country.get("population"),
        "Area": country.get("area"),
        "Currency": ', '.join(country.get("currencies", {}).keys()) if country.get("currencies") else None,
    }
    data_rows.append(row)



```

```python
import pandas as pd
df=pd.DataFrame(data_rows)
df.to_csv("All_countries_dataset.csv",index = False)


```

```python
import requests
import pandas as pd

# fetch data from the API

url = "https://fakestoreapi.com/products"
response = requests.get(url)
products = response.json()


# convert to data frame

df = pd.DataFrame(products)
df.head()
```

```python
df['rating_rate']=df['rating'].apply(lambda x:x ['rate'])
df['rating_count']=df['rating'].apply(lambda x:x ['count'])

```

```python
# display

df[["title","price","category","description","image","rating_rate","rating_count"]].head()
```

```python
df['Discount_price']=df['price']*0.9
df
```

```python
df['is_high_rate']=df['rating_rate']>=4
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(y="category",data = df)
plt.title ("No. Of product per category ")
plt.show()
```
